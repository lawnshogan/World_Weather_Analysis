```python
# Download the donors2008.csv.
# Open and read the donors2008.csv file into a pandas DataFrame.
# Using groupby, create a Series with the average “Amount” for each state.
# Convert the Series to a DataFrame if needed.
# How many states have averages equal to or greater than $750.00?

import pandas as pd
donor_data = "../Python_Unit_Assessment/donors2008.csv"
```

```python
donor_data_df = pd.read_csv(donor_data, encoding="ISO-8859-1")
donor_data_df.head(10)
```

```python
amount_df = donor_data_df.groupby(['State','Amount']).mean()

amount_df
```
