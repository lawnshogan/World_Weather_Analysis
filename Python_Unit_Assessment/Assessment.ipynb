```python
# Download the donors2008.csv.
# Open and read the donors2008.csv file into a pandas DataFrame.
# Using groupby, create a Series with the average “Amount” for each state.
# Convert the Series to a DataFrame if needed.
# How many states have averages equal to or greater than $750.00?

import pandas as pd
import numpy as np

donor_data = "../Python_Unit_Assessment/donors2008.csv"
```

```python
donor_data_df = pd.read_csv(donor_data, encoding="ISO-8859-1")
donor_data_df.head(10)
```

```python

amount_df = donor_data_df.groupby(['State','Amount']).mean()
amount_df.head(100)
```

```python
amount_averages = donor_data_df.groupby('State').Amount.mean() >= 750



amount_averages
```

```python
# Bins
test_grades = {
    'Class': ['Oct', 'Oct', 'Jan', 'Jan', 'Oct', 'Jan'], 
    'Name': ["Cyndy", "Logan", "Laci", "Elmer", "Crystle", "Emmie"], 
    'Test Score': [90, 59, 72, 88, 98, 60]}

test_grades_df = pd.DataFrame(test_grades)
test_grades_df.head()
```

```python
bins = [0, 59, 69, 79, 89, 100]
group_names = ["F", "D", "C", "B", "A"]
```

```python
# Add Grade column and assign grade based on Test score
test_grades_df["Grade"] = pd.cut(test_grades_df["Test Score"], bins, labels=group_names)
test_grades_df.head(10)
```
